```python
import pandas as pd
import numpy as np
foodDf = pd.read_excel('../data/raw_extract/furtherExtracted/FoodPhoto Event-PortionsPercentEaten.xlsx')
forgotDf = pd.read_excel('../data/raw_extract/furtherExtracted/ForgotMeal Event-Portions.xlsx')
```

```python
foodDfCols = ["SubjectID","DateAsString", "DateTime", "Date", "Time",
              "Meal", "FoodDictionaryIndex","Food", "PortionMultiplier",
              "PercentEaten","PortionTypeTemplate","AddedSugar","Alcohol",
              "Dairy", "Fruit","Grains","GramWt","KCals",
              "Protein", "SatFat", "Vegetable"]

foodDf['DateTime'] = pd.to_datetime(foodDf['DateAsString'])
foodDf['Date']=foodDf['DateTime'].dt.date
foodDf['Time']=foodDf['DateTime'].dt.time
foodDf = foodDf.rename({"AboutSubjectID":"SubjectID",
                        "MealWhatMeal":"Meal",
                        "TemplateAddedSugar":"AddedSugar",
                        "TemplateAlcohol":"Alcohol",
                        "TemplateDairy":"Dairy",
                        "TemplateFruit":"Fruit",
                        "TemplateGrains":"Grains",
                        "TemplateGramWt":"GramWt",
                        "TemplateKCals":"KCals",          
                        "TemplateProtein":"Protein",
                        "TemplateSatFat":"SatFat",
                        "TemplateVegetable":"Vegetable"},axis=1)
```

```python
anaCols = ["AddedSugar","Alcohol", "Dairy", "Fruit","Grains","GramWt","KCals", "Protein", "SatFat", "Vegetable"]
foodDf[anaCols] = foodDf[anaCols].multiply(other=foodDf["PercentEaten"],axis=0)
anaFoodDf = foodDf[foodDfCols]
```

```python
forgotTempDf = pd.read_excel('../data/raw_extract/furtherExtracted/ForgotMeal Event-Portions.xlsx')
forgotTempDf['DateTime'] = pd.to_datetime(forgotDf['DateAsString'])
forgotTempDf['Date'] = forgotTempDf['DateTime'].dt.date
forgotTempDf['Time'] = forgotTempDf['DateTime'].dt.time
forgotTempDf["PercentEaten"] = 1
forgotTempDf = forgotTempDf.rename({"AboutSubjectID":"SubjectID",
                        "MealWhatMeal":"Meal",
                        "TemplateAddedSugar":"AddedSugar",
                        "TemplateAlcohol":"Alcohol",
                        "TemplateDairy":"Dairy",
                        "TemplateFruit":"Fruit",
                        "TemplateGrains":"Grains",
                        "TemplateGramWt":"GramWt",
                        "TemplateKCals":"KCals",          
                        "TemplateProtein":"Protein",
                        "TemplateSatFat":"SatFat",
                        "TemplateVegetable":"Vegetable"},axis=1)

forgotTempDfFilt = forgotTempDf['PortionTypeTemplate'] == 1
anaForgotTempDf = forgotTempDf[forgotTempDfFilt][foodDfCols]
```

```python
forgotStdDf = pd.read_excel('../data/raw_extract/furtherExtracted/ForgotMeal Event-Portions.xlsx')
forgotStdDf['DateTime'] = pd.to_datetime(forgotStdDf['DateAsString'])
forgotStdDf['Date'] = forgotStdDf['DateTime'].dt.date
forgotStdDf['Time'] = forgotStdDf['DateTime'].dt.time
forgotStdDf["PercentEaten"] = 1
forgotStdDf = forgotStdDf.rename({"AboutSubjectID":"SubjectID",
                        "MealWhatMeal":"Meal",
                        "StdServingAddedSugar":"AddedSugar",
                        "StdServingAlcohol":"Alcohol",
                        "StdServingDairy":"Dairy",
                        "StdServingFruit":"Fruit",
                        "StdServingGrains":"Grains",
                        "StdServingGramWt":"GramWt",
                        "StdServingKCals":"KCals",          
                        "StdServingProtein":"Protein",
                        "StdServingSatFat":"SatFat",
                        "StdServingVegetable":"Vegetable"},axis=1)
forgotStdFilt = forgotStdDf['PortionTypeTemplate'] == 0
anaStdForgotDf = forgotStdDf[forgotStdFilt][foodDfCols]
```

```python
anaForgotDf = anaStdForgotDf.append(anaForgotTempDf)
anaForgotDf[anaCols] = anaForgotDf[anaCols].multiply(other=anaForgotDf["PortionMultiplier"],axis=0)
perFoodDf = anaForgotDf.append(anaFoodDf)
perFoodDf['FoodID'] = (perFoodDf['SubjectID'].map(str) + '_' +
                       perFoodDf['Date'].map(str) + '_' + 
                       perFoodDf['Meal'].map(str) + '_' +
                       perFoodDf['FoodDictionaryIndex'].map(str))
perFoodDf = perFoodDf.set_index('FoodID')
```

```python
perFoodDf = anaForgotDf.append(anaFoodDf)
perFoodDf['MealID'] = (perFoodDf['SubjectID'].map(str) + '_' +
                       perFoodDf['Date'].map(str) + '_' + 
                       perFoodDf['Meal'].map(str))
meal_group = perFoodDf.groupby('MealID')
perMealDf = meal_group[anaCols].agg('sum')
perMealDfCols = ["SubjectID", "Date", "Meal", "MealID"]
mergeWithPerMeal = perFoodDf[perMealDfCols].set_index('MealID').drop_duplicates()
perMealDf = mergeWithPerMeal.join(perMealDf)
```

```python
perFoodDf = anaForgotDf.append(anaFoodDf)
perFoodDf['DayID'] = perFoodDf['SubjectID'].map(str) + '_' + perFoodDf['Date'].map(str)
day_group = perFoodDf.groupby('DayID')
perDayDf = day_group[anaCols].agg('sum')
perDayDfCols = ["SubjectID", "Date", "DayID"]
mergeWithPerDay = perFoodDf[perDayDfCols].set_index('DayID').drop_duplicates()
perDayDf = mergeWithPerDay.join(perDayDf)
```
